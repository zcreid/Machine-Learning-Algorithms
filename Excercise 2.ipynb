````python
# 'price' has already been created
discount = {-1:1, 0:.95, 100:.9, 200: .8, 300: .7}

# Write your code here
def discounted_price(price):
  total = 0
  for key in discount:
    if price >= key:
      total = price * discount[key]
  return total

# Example usage
print(discounted_price(250))  # Output: 200.0
print(discounted_price(50))   # Output: 47.5
````

*   If the price is between `100` and `200` (`100` inclusive), there will be a `10%` discount.

````python
# 'price' has already been created
discount = {-1:1, 0:.95, 100:.9, 200: .8, 300: .7}

# Write your code here
def discounted_price(price):
  total = 0
  for key in discount:
    if price >= key:
      total = price * discount[key]
  return total

# Example usage
print(discounted_price(250))  # Output: 200.0
print(discounted_price(50))   # Output: 47.5
````
